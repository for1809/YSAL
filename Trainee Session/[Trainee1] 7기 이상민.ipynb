```python
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
from bs4 import BeautifulSoup


```

### 크롬 드라이버 설정

```python
options = Options()
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')

# headless 모드로 하고 싶으신 경우 아래 줄의 주석 처리를 해제 (# 삭제)하시면 됩니다.
# options.add_argument('--headless') 

service = Service(ChromeDriverManager().install()) 

# 크롬드라이버 실행
driver = webdriver.Chrome(service = service, options = options)
```

## 1.  기사 제목 스크랩

### 기사 스크랩 함수

```python
# 기사 스크랩하는 과정을 함수로 만들어주기
def scrap_news_title(driver): 

    # 제목에 해당하는 요소 선택 -> F12로 확인
    headline_spans = driver.find_elements(By.CSS_SELECTOR, "span.sds-comps-text-type-headline1") 
    # 리스트에 저장
    article_titles = [span.text.strip() for span in headline_spans]

    return article_titles
```

### 크롤링에 필요한 동작 구현

0000 으로 되어있는 부분을 요구사항에 맞게 채워주시면 됩니다. 

```python
#크롬 드라이버에 url 주소 넣고 실행
driver.get('https://www.naver.com/')

# 페이지가 완전히 로딩되도록 3초동안 기다림
time.sleep(3)
```

```python
# 검색어 창을 찾아 search_box 변수에 저장 (By.XPATH 방식)
search_box = driver.find_element(By.XPATH, '//input[@type="search" and @name="query" and @id="query"]')
# 힌트: 입력을 의미하는 태그를 찾아보세요!
```

```python
search_box.click()    # 객체 클릭 (여기서는 검색어 창)
search_box.clear()    # 혹시 검색창에 이전 값이 있으면 지우고
search_box.send_keys("구자욱", Keys.RETURN) # 검색어 선택 후 클릭 (원하시는 검색어를 넣어보세요!)
time.sleep(1)
```

```python
# (기존 검색 클릭 후) 결과를 로딩할 시간 확보
from selenium.webdriver.support.ui import WebDriverWait
wait = WebDriverWait(driver, 10)
news_button = wait.until(EC.element_to_be_clickable((By.LINK_TEXT, '뉴스')))
news_button.click()
```

```python
titles = [] # 기사 제목을 저장하기 위한 리스트 생성
if __name__ == "__main__":

    # 현재 켜져있는 url을 변수로 저장 -> 이 사이트에서 기사 제목을 스크랩해오겠다. 
    url = driver.current_url 

    # idx: 순서 인덱스를 받아오기
    # title: scrap_news_title를 실행해서 받아온 제목

    for j in range(5): # 반복 횟수 지정 -> 스크롤을 내릴 횟수를 정해주세요!
        for idx, title in enumerate((scrap_news_title(driver)), 1): 
            print(f"{idx}. {title}")
            titles.append(title) # titles 리스트에 저장
        # 해당 페이지의 기사제목을 다 모았다면 맨 아래로 스크롤
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
```

### 수집한 기사 제목을 DataFrame으로 만들기


```python
# titles 리스트에 저장한 값을 DataFrame으로 변환
df_headlines = pd.DataFrame({
    "headline": titles
})
```

```python
# 엑셀 혹은 csv로 내보내기
df_headlines.to_csv("news_headlines.csv", index=False)
```

## 2. Statiz의 기록실에서 지표 수집하기

```python
# (채우기) 기록실 버튼 클릭
# 힌트: 뉴스 버튼 클릭...!!
archive = driver.find_element(By.XPATH, '//span[@class="text"]')
archive.click()
```

```python
# (채우기) 투구 버튼 클릭
wait = WebDriverWait(driver, 10)
pitching = wait.until(EC.element_to_be_clickable((By.XPATH, '//button[contains(text(), "투구")]')))
pitching.click()
```

```python
# (채우기) 테이블의 xpath 지정하기
table = driver.find_element(By.XPATH, '//div[@class="table_type01 transverse_scroll cbox"]')
```

```python
#get_attribute 함수는 element 안의 html 내용들을 모두 가져온다
table_html = table.get_attribute('outerHTML')
bsObject = BeautifulSoup(table_html, 'html.parser')
```

```python
column = ['Name', 'WAR', 'G', 'ERA', 'FIP', 'WHIP'] # 테이블에서 추출할 칼럼명을 "name", "team", ... 과 같이 적어주세요!
rows = []

#html 문서에서 행을 의미하는 태그(tr) 작성
bs_len = len(bsObject.find_all("tr"))

for i in range(2, bs_len):
    bsTr = bsObject.find_all("tr")[i]

    #html문서에서 열 이름을 의미하는 태그(th) 작성
    if bsTr.find("th") is not None:
        continue

    #html문서에서 셀 데이터를 의미하는 태그(td) 작성
    tds = bsTr.find_all("td")


    #tds[idx].text.strip()는 데이터를 텍스트로 추출하고, 공백 제거
    row = {col: tds[idx].text.strip() for idx, col in enumerate(column)}

    #빈 리스트에 row 채워넣기
    rows.append(row)
print(rows)
df_pitch = pd.DataFrame(rows)
df_pitch
# (채우기) df_pitch에 rows를 데이터프레임 형태로 채워보세요!
```
